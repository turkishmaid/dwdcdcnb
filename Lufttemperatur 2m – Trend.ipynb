# Lufttemperatur 2m – Trend

Datenquelle sind die Stundenwerte, so dass man mit Mittagstemperatur, tropischen Nächten usw. spielen kann.

```python
# initialize interactive johanna
import johanna
johanna.interactive(dotfolder="~/.dwd-cdc", dbname="hr-temp.sqlite")

# initialize matplotlib
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# import dwdcdc.at2h
from dwdcdc.interactive import trend, metatrend
from dwdcdc.const import MANNHEIM, POTSDAM
```

## Der Juni in Mannheim

```python
trend.plot(plt, station=MANNHEIM, monat=6, stunde=12, von=0, bis=3000)
```

## Der Juni in Potsdam

### 1893 - 2020

```python
trend.plot(plt, station=POTSDAM, monat=6, stunde=12, von=0, bis=3000)
```

### 1948 - 2020

Für bessere Vergleichbarkeit mit Mannheim

```python
trend.plot(plt, station=POTSDAM, monat=6, stunde=12, von=1948, bis=3000)
```

### Wie ändert sich die Trendline?

Jetzt verkürzen wir den Zeitraum, über den der Trend gebildet wird. Natürlich in 11-er Schritten, um der Sonnenaktivität Rechnung zu tragen. Der Juni ist eindrucksvoll, aber der August ist erschreckend...

#### Juni

```python
metatrend.plot(plt, station=POTSDAM, monat=6, stunde=12, von=0, bis=3000)
```

#### August

```python
metatrend.plot(plt, station=POTSDAM, monat=8, stunde=12, von=0, bis=3000)
```
